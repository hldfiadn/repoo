# Eksplorasi Train

## informasi data - train

```python
df_train = pd.read_csv('/content/drive/MyDrive/PECEDE/tugas akhir/train/_classes.csv')
```

```python
df_train
```

```python
df_train.info()
```

```python
df_train.describe()
```

## visualisasi distribusi kategori - train

```python
df_train.columns
```

```python
#Menghitung jumlah citra per kategori
category_counts_train = df_train[[' apple', ' avocado', ' capsicum', ' mango', ' orange',
       ' rockmelon', ' strawberry']].sum()

#Membuat plot bar
category_counts_train.plot(kind='bar')
plt.title('Distribusi Kategori')
plt.xlabel('Kategori')
plt.ylabel('Jumlah Citra')
plt.show()
```

```python
category_counts_train
```

```python
data_for_heatmap = df_train[[' apple', ' avocado', ' capsicum', ' mango', ' orange',
       ' rockmelon', ' strawberry']]

#Membuat heatmap
plt.figure(figsize=(10,8))  #Menyesuaikan ukuran gambar
sns.heatmap(data_for_heatmap, annot=True, cmap='coolwarm', cbar=True, fmt='g')
plt.title('Heatmap Distribusi Kategori')
plt.xlabel('Kategori')
plt.ylabel('Indeks')
plt.show()
```

```python
data_for_visual = df_train[[' apple', ' avocado', ' capsicum', ' mango', ' orange',
       ' rockmelon', ' strawberry']]

# Mengatur ukuran gambar
plt.figure(figsize=(12, 6))

# Menambahkan judul dan label
plt.title('Distribusi Kategori')
plt.xlabel('Kategori')
plt.ylabel('Indeks')

# Mengatur tanda pada sumbu x agar menunjukkan nama kategori
plt.xticks(ticks=np.arange(data_for_visual.shape[1]), labels=data_for_visual.columns)
plt.yticks([])  # Hapus label pada sumbu y jika tidak diperlukan

# Mengiterasi melalui setiap indeks dan kategori untuk menampilkan kotak berwarna
for index, row in data_for_visual.iterrows():
    for col_index, value in enumerate(row):
        if value == 1:
            plt.fill_betweenx([index - 0.5, index + 0.5], col_index - 0.5, col_index + 0.5, color='#377eb8')

# Menampilkan grid untuk memperjelas batas
plt.grid(True)

# Menampilkan plot
plt.show()
```
