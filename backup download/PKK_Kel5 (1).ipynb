# Fuzzification

Sterilising and threshing station

```python
# Create global variables
# Input initialization
x_12 = np.arange(0, 20, 1) # The percentage of unripe fruits
x_22 = np.arange(70, 150.5, 10) # #Sterilising time
x_32 = np.arange(2.5, 3.300000000000001, 0.1) # Steam pressure

# Output initialization
ow_12 = np.arange(0, 0.60001, 0.1) #. % of oil loss in condensate water
ow_13 = np.arange(0, 4.001, 0.5) # % of oil loss in empty bunches
ow_23 = np.arange(0, 1.201, 0.01) # % of oil loss in the fruits remaining in empty bunches

# Create fuzzy membership functions
# Input
x12_low = fuzz.trapmf(x_12, [0, 0, 3, 9])
x12_high = fuzz.trapmf(x_12, [3, 9, 19, 19])

x22_short = fuzz.trapmf(x_22, [70,70, 90, 110])
x22_medium = fuzz.trimf(x_22, [90, 110, 130])
x22_long = fuzz.trapmf(x_22, [110, 130, 150, 150 ])

x32_low = fuzz.trapmf(x_32, [2.5,2.5, 2.7, 2.9])
x32_medium = fuzz.trimf(x_32, [2.7, 2.9, 3.1])
x32_high = fuzz.trapmf(x_32, [2.9, 3.1, 3.4, 3.4 ])

# Output
ow12_low = fuzz.trapmf(ow_12, [0, 0, 0.2, 0.3])
ow12_medium = fuzz.trimf(ow_12, [0.2, 0.3, 0.4])
ow12_high = fuzz.trapmf(ow_12, [0.3, 0.4, 3.3, 3.3 ])

ow13_low = fuzz.trapmf(ow_13, [0, 0, 1.5, 2])
ow13_medium = fuzz.trimf(ow_13, [1.5, 2, 2.5])
ow13_high = fuzz.trapmf(ow_13, [2, 2.5, 4.1, 4.1 ])

ow23_low = fuzz.trapmf(ow_23, [0, 0, 0.2, 0.35])
ow23_medium = fuzz.trimf(ow_23, [0.2, 0.35, 0.5])
ow23_high = fuzz.trapmf(ow_23, [0.35, 0.5, 1.3, 1.3 ])
```

```python
# Visualization of membership functions
fig, (ax0, ax1, ax2, ax3, ax4, ax5) = plt.subplots(nrows=6, figsize=(14,20))

ax0.plot(x_12, x12_low, 'b',   linewidth=1.5,label='Low')
ax0.plot(x_12, x12_high, 'r',  linewidth=1.5, label='High')
ax0.set_title('The percentage of unripe fruits (X12)')
ax0.legend()

ax1.plot(x_22, x22_short, 'b', linewidth=1.5, label='Short')
ax1.plot(x_22, x22_medium, 'g', linewidth=1.5, label='Medium')
ax1.plot(x_22, x22_long, 'r', linewidth=1.5, label='Long')
ax1.set_title(' Sterilising time (X22)')
ax1.legend()

ax2.plot(x_32, x32_low, 'b', linewidth=1.5, label='Low')
ax2.plot(x_32, x32_medium, 'g', linewidth=1.5, label='Medium')
ax2.plot(x_32, x32_high, 'r', linewidth=1.5, label='High')
ax2.set_title('Steam pressure (X32)')
ax2.legend()

ax3.plot(ow_12, ow12_low, 'b', linewidth=1.5, label='Low')
ax3.plot(ow_12, ow12_medium, 'g', linewidth=1.5, label='Medium')
ax3.plot(ow_12, ow12_high, 'r', linewidth=1.5, label='High')
ax3.set_title('The percentage of oil in condensate water (OW12)')
ax3.legend()

ax4.plot(ow_13, ow13_low, 'b', linewidth=1.5, label='Low')
ax4.plot(ow_13, ow13_medium, 'g', linewidth=1.5, label='Medium')
ax4.plot(ow_13, ow13_high, 'r', linewidth=1.5, label='High')
ax4.set_title('The percentage of oil in empty bunches  (OW13)')
ax4.legend()

ax5.plot(ow_23, ow23_low, 'b', linewidth=1.5, label='Low')
ax5.plot(ow_23, ow23_medium, 'g', linewidth=1.5, label='Medium')
ax5.plot(ow_23, ow23_high, 'r', linewidth=1.5, label='High')
ax5.set_title('The percentage of oil in the fruit remaining in the empty bunches (OW23)')
ax5.legend()

# Turn off top/right axes
for ax in (ax0, ax1, ax2, ax3, ax4, ax5):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.tight_layout()
```

```python
# =============================================================================
# Rule definition and Membership degree
# =============================================================================

# Membership degree of unripe fruit
x12_level_low = fuzz.interp_membership(x_12, x12_low, 16)
x12_level_high = fuzz.interp_membership(x_12, x12_high, 16)

x12_fuzz_var = np.array([x12_level_low, x12_level_high])
print('Fuzzy Variable(X12 =' , 16, ') :', x12_fuzz_var)

# Membership degree of sterilising time
x22_level_short = fuzz.interp_membership(x_22, x22_short, 140)
x22_level_medium = fuzz.interp_membership(x_22, x22_medium, 140)
x22_level_long = fuzz.interp_membership(x_22, x22_long, 140)

x22_fuzz_var = np.array([x22_level_short , x22_level_medium, x22_level_long])
print('Fuzzy Variable(X22 =' , 140, ') :', x22_fuzz_var)

# Membership degree of steam pressure
x32_level_low = fuzz.interp_membership(x_32, x32_low,3.2)
x32_level_medium = fuzz.interp_membership(x_32, x32_medium, 3.2)
x32_level_high = fuzz.interp_membership(x_32, x32_high, 3.2)

x32_fuzz_var = np.array([x32_level_low, x32_level_medium, x32_level_high])
print('Fuzzy Variable(X32 =' , 3.2, ') :', x32_fuzz_var)

# Rules definition
rule1_1 = np.fmin(x12_level_high, np.fmin(x22_level_long, x32_level_high))
rule1_2 = np.fmin(x12_level_high, np.fmin(x22_level_long, x32_level_medium))
rule1_3 = np.fmin(x12_level_high, np.fmin(x22_level_long, x32_level_low))
rule1_4 = np.fmin(x12_level_high, np.fmin(x22_level_medium, x32_level_high))
rule1_5 = np.fmin(x12_level_high, np.fmin(x22_level_medium, x32_level_medium))
rule1_6 = np.fmin(x12_level_high, np.fmin(x22_level_medium, x32_level_low))
rule1_7 = np.fmin(x12_level_high, np.fmin(x22_level_short, x32_level_high))
rule1_8 = np.fmin(x12_level_high, np.fmin(x22_level_short, x32_level_medium))
rule1_9 = np.fmin(x12_level_high, np.fmin(x22_level_short, x32_level_low))
rule1_10 = np.fmin(x12_level_low, np.fmin(x22_level_long, x32_level_high))
rule1_11 = np.fmin(x12_level_low, np.fmin(x22_level_long, x32_level_medium))
rule1_12 = np.fmin(x12_level_low, np.fmin(x22_level_long, x32_level_low))
rule1_13 = np.fmin(x12_level_low, np.fmin(x22_level_medium, x32_level_high))
rule1_14 = np.fmin(x12_level_low, np.fmin(x22_level_medium, x32_level_medium))
rule1_15 = np.fmin(x12_level_low, np.fmin(x22_level_medium, x32_level_low))
rule1_16 = np.fmin(x12_level_low, np.fmin(x22_level_short, x32_level_high))
rule1_17 = np.fmin(x12_level_low, np.fmin(x22_level_short, x32_level_medium))
rule1_18 = np.fmin(x12_level_low, np.fmin(x22_level_short, x32_level_low))
```
